```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

## Section Ø¢

```python
lower_limit = cv.imread('./background-LE.bmp', 0)
upper_limit = cv.imread('./fullscale-LE.bmp', 0)
raw_image = cv.imread('./object-LE.bmp', 0)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 3))
ax[0].imshow(raw_image, cmap='gray',vmin=0, vmax=255)
ax[0].set_title('Raw Image')
ax[1].imshow(upper_limit, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Upper Limit Image')
ax[2].imshow(lower_limit, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('Lower Limit Image')
plt.show()
```

## Section Ø¨

```python
rowAvg_upper_limit = np.mean(upper_limit, axis=1, keepdims=True)
rowAvg_lower_limit = np.mean(lower_limit, axis=1, keepdims=True)
```

## Section Ø¬

```python
normalized_image = (raw_image - rowAvg_lower_limit) / (rowAvg_upper_limit - rowAvg_lower_limit)
normalized_image *= 255
```

## Section Ø¯

```python
fix, ax = plt.subplots(1, 2, figsize=(16, 8))
ax[0].imshow(raw_image, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Raw Image')
ax[1].imshow(normalized_image, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Normalized Image')
plt.show()
```
